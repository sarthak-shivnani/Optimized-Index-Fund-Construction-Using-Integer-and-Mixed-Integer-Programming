```python

def filling_null_values(df):
    '''2023data.csv has some null values in the 'NDX' column. 
    This function fills those null values with actual values from yahoo finance.
    Also it replaces the null values in the 'ARM' column with 0 for rows where 'Date' is less than 2023-09-14 since it was not listed before that date.
    '''
    #replacing the null values in the data with actual values from web
    #replacing the null values in the data with actual values from web
    stock_price_correction_list = [('2023-03-10',11830.28),('2023-06-12',14784.30),('2023-06-14',15005.69),('2023-06-21',14867.45),('2023-07-06',15089.45),('2023-11-22',16001.39)]
    for correction_date,correction_price in stock_price_correction_list:
        df.loc[df['Date'] == correction_date, 'NDX'] = correction_price

    # Ensure the 'Date' column is in datetime format for accurate comparisons
    # stocks_2023_raw['Date'] = pd.to_datetime(stocks_2023_raw['Date'])

    # Fill NaNs in 'ARM' column with 0 for rows where 'Date' is less than 2023-09-14 since it was not listed before that date
    df.loc[df['Date'] < '2023-09-14', 'ARM'] = df.loc[df['Date'] < '2023-09-14', 'ARM'].fillna(0)
    return df

```

```python
def interpolate_values(df):
    """
    Interpolate missing values in a dataframe using linear interpolation and handles nulls/infs.
    """
    df_interpolated = df.interpolate(method='linear', limit_direction='both')
    if df_interpolated.isnull().values.any():
        df_interpolated = df_interpolated.fillna(method='ffill').fillna(method='bfill')
    return df_interpolated
```

```python
def get_return_matrix(df):
    '''gets daily returns of the stocks in the dataframe and replaces inf values with 0'''
    returns_df = df.copy()  # Create a copy of the original DataFrame
    returns_df.drop(columns=['Date'], inplace=True)  # Drop the 'Date' column

    # Calculate daily returns
    returns_df = returns_df.pct_change().fillna(0)
    returns_df.replace([np.inf, -np.inf], 0, inplace=True)
    return returns_df
```

```python
def get_correlation_df(df):
    '''
    This function takes in a dataframe and returns a Pearson correlation matrix of the features.
    '''
    df = df.drop(columns=['NDX']).corr()
    return df
```

### Q2 - Stock Selection

Now that we have the correlation matrix for the returns we can build an optimization problem to pick m stocks that maximizes the correlation between the selected stock and representative from the NASDAQ index.

```python


def select_similar_stocks(m, correl_2023):
    n = correl_2023.shape[0]  # Number of stocks based on the correlation matrix size
    
    # Initialize the model
    fund_creator = gp.Model()
    selected_stocks = fund_creator.addMVar(n, vtype='B')  # Binary variable for stock selection
    linking_stocks = fund_creator.addMVar((n, n), vtype='B')  # Binary variable for stock linking

    # Add constraints
    selection_const = fund_creator.addConstr(gp.quicksum(selected_stocks[i] for i in range(n)) <= m)
    one_representative_const = fund_creator.addConstrs(
        gp.quicksum(linking_stocks[i, j] for j in range(n)) == 1 for i in range(n)
    )
    ensure_best_representative_const = fund_creator.addConstrs(
        linking_stocks[i, j] <= selected_stocks[j] for j in range(n) for i in range(n)
    )

    # Set the objective function
    fund_creator.setObjective(
        gp.quicksum(correl_2023.iloc[i, j] * linking_stocks[i, j] for i in range(n) for j in range(n)),
        sense=gp.GRB.MAXIMIZE
    )
    
    # Suppress output and optimize
    fund_creator.Params.OutputFlag = 0
    fund_creator.optimize()

    # Extract selected stock indices
    selection_series = pd.Series(selected_stocks.x)
    selected_columns = correl_2023.columns[selection_series == 1].tolist()

    # Calculate the average total similarity
    avg_similarity = fund_creator.objVal / n
    linking_stocks = pd.DataFrame(linking_stocks.x.tolist())

    # Print or return results
    print("Selected Stocks:", selected_columns)
    print("\nMax average total similarity (correlation observed):", avg_similarity)

    return linking_stocks,selected_columns, avg_similarity
```

### Q2 - Portfolio Weights calculation

Now let's identify how much of each selected stock should we have in our fund. This is identified by optimizing to achieve minimum difference between the NASDAQ index and weighted returns from the selected stocks. The weights here represent the % of stock in the fund.

```python
def calculate_portfolio_weights(num_days_2023, fund_stocks, nasdaq_index_2023, selected_stock_names, plot_bar_chart=True):
    num_selected_stocks = len(selected_stock_names)  # Number of stocks selected in the fund

    # Initializing the model
    portfolio_model = gp.Model()
    
    # Adding decision variables
    fund_weights = portfolio_model.addMVar(num_selected_stocks)
    index_differences = portfolio_model.addMVar(num_days_2023)

    # Adding constraints
    portfolio_model.addConstr(gp.quicksum(fund_weights[i] for i in range(num_selected_stocks)) == 1)
    
    portfolio_model.addConstrs(
        (gp.quicksum(fund_weights[i] * fund_stocks.iloc[day, i] for i in range(num_selected_stocks)) - nasdaq_index_2023[day]) <= index_differences[day]
        for day in range(num_days_2023)
    )

    portfolio_model.addConstrs(
        (-1 * gp.quicksum(fund_weights[i] * fund_stocks.iloc[day, i] for i in range(num_selected_stocks)) + nasdaq_index_2023[day]) <= index_differences[day]
        for day in range(num_days_2023)
    )

    # Set the objective function
    portfolio_model.setObjective(gp.quicksum(index_differences[i] for i in range(num_days_2023)), sense=gp.GRB.MINIMIZE)
    
    # Suppress output and optimize
    portfolio_model.Params.OutputFlag = 0
    portfolio_model.optimize()

    # Extract the results into a DataFrame
    fund_weights_df = pd.DataFrame({
        'stock': selected_stock_names,
        'weights': fund_weights.X.tolist()
    })
    
    # Sorting the DataFrame by weights in descending order
    fund_weights_df = fund_weights_df.sort_values('weights', ascending=False)

    # Print the objective value and return the DataFrame
    print('\nThe difference between 2023 NASDAQ Index return and our fund return =', round(portfolio_model.objVal, 2))
    print('\nThe weights assigned to stocks in our fund are:')

    if plot_bar_chart:
        display(fund_weights_df)
        # Plot the data as a bar graph
        plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
        plt.bar(fund_weights_df['stock'], fund_weights_df['weights'])
        plt.xlabel('Stock Name')
        plt.ylabel('Stock Weight')
        plt.title('Weight Distribution of the Stocks in Our Fund')
        plt.show()
    
    return fund_weights_df, round(portfolio_model.objVal, 2)
```

```python
def calculate_and_evaluate_returns_2024(stock_data_2024, selected_columns, portfolio_weights_df, training_error, plot_daily_returns=True):
    # Calculate daily returns for 2024
    daily_returns_2024 = stock_data_2024.copy()  # Create a copy of the original DataFrame
    daily_returns_2024.drop(columns=['Date'], inplace=True, errors='ignore')  # Drop 'Date' column if it exists
    daily_returns_2024 = daily_returns_2024.pct_change().fillna(0)  # Calculate daily returns and handle NaNs
    index_returns_2024 = daily_returns_2024['NDX']  # Extract NASDAQ-100 index returns for 2024

    # Filter 2024 returns for selected stocks
    filtered_returns_2024 = daily_returns_2024[selected_columns]
    portfolio_weights = portfolio_weights_df['weights']
    
    # Calculate weighted returns for the selected stocks fund
    weighted_returns_2024 = pd.DataFrame()
    for i, column in enumerate(filtered_returns_2024.columns):
        weighted_returns_2024[column] = filtered_returns_2024[column] * portfolio_weights[i]
    weighted_returns_2024['total_weighted_returns'] = weighted_returns_2024[selected_columns].sum(axis=1)
    
    # Calculate testing error
    testing_error_2024 = sum(abs(index_returns_2024 - weighted_returns_2024['total_weighted_returns']))

    # Print training and testing errors
    print('\nThe difference between the 2023 NASDAQ Index and the fund we have created (Training error) = ', round(training_error, 2), '\n')
    print('The difference between the 2024 NASDAQ Index and the fund we have created (Testing error) = ', round(testing_error_2024, 2))

    # Plotting if the flag is set to True
    if plot_daily_returns:
        plt.figure(figsize=(12, 6))
        plt.plot(index_returns_2024.index, index_returns_2024, label='NASDAQ Index (NDX)', color='blue', linestyle='-')  # Solid line for index
        plt.plot(weighted_returns_2024.index, weighted_returns_2024['total_weighted_returns'], label='Selected Stocks Fund', color='orange', linestyle='--')  # Dotted line for fund
        plt.title("Daily Returns in 2024: NASDAQ Index vs. Selected Stocks Fund")
        plt.xlabel("Days")
        plt.ylabel("Daily Return")
        plt.legend()
        plt.show()

    return testing_error_2024, daily_returns_2024


```

#### Let's find out how the error changes as we increase the value of m

```python
def calculate_training_and_testing_errors(m_values, stock_selection_dict, weights_dict, returns_2023_df, returns_2024_df, selection_objective):
    training_errors_2023 = []
    testing_errors_2024 = []

    for m in m_values:
        # Training error calculation for 2023
        filtered_returns_2023 = returns_2023_df[stock_selection_dict[m]]
        weights = weights_dict[m]['weights'].tolist()
        
        weighted_returns_2023 = pd.DataFrame()
        for i, column in enumerate(filtered_returns_2023.columns):
            weighted_returns_2023[column] = filtered_returns_2023[column] * weights[i]
        weighted_returns_2023['total_weighted_returns'] = weighted_returns_2023[stock_selection_dict[m]].sum(axis=1)
        
        training_error = np.sum(abs(returns_2023_df['NDX'] - weighted_returns_2023['total_weighted_returns']))
        training_errors_2023.append(training_error)

        # Testing error calculation for 2024
        filtered_returns_2024 = returns_2024_df[stock_selection_dict[m]]
        
        weighted_returns_2024 = pd.DataFrame()
        for i, column in enumerate(filtered_returns_2024.columns):
            weighted_returns_2024[column] = filtered_returns_2024[column] * weights[i]
        weighted_returns_2024['total_weighted_returns'] = weighted_returns_2024[stock_selection_dict[m]].sum(axis=1)
        
        testing_error = np.sum(abs(returns_2024_df['NDX'] - weighted_returns_2024['total_weighted_returns']))
        testing_errors_2024.append(testing_error)

    # Creating the results DataFrame
    errors_df = pd.DataFrame({
        'selected_stocks': m_values,
        'stock_selection_objective': selection_objective,
        '2023_errors': training_errors_2023,
        '2024_errors': testing_errors_2024
    })

    errors_df.set_index('selected_stocks', inplace=True)
    
    return errors_df
```
